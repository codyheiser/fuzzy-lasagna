```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(style = 'white')
%matplotlib inline
```

```python
img = cv2.imread('inputs/IHC_RegisteredToPreAF_pyramid.ome (d=4.00, x=12000, y=12000, w=4000, h=4000, z=0).tif') # 1 = read as color image
```

```python
plt.figure(figsize=(7,7))
plt.imshow(img[:,:,2])
plt.tick_params(labelbottom=False, labelleft=False)
plt.show()
```

```python
blur = cv2.GaussianBlur(img,(5,5),0)
plt.figure(figsize=(7,7))
plt.imshow(blur[:,:,2])
plt.tick_params(labelbottom=False, labelleft=False)
plt.show()
```

```python
ret, thresh = cv2.threshold(img[:,:,2],0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
```

```python
plt.figure(figsize=(7,7))
plt.imshow(thresh)
plt.tick_params(labelbottom=False, labelleft=False)
plt.show()
```
